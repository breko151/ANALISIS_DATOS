### Lectura CSV

Hacemos uso de la librería CSV

```python
import csv
```

```python
path = 'C:/Users/USUARIO DELL/Documents/Python Scripts/ANALISIS_DATOS/LECTURA_FICHEROS/subvenciones-educacion2015.csv'
with open(path, encoding = 'latin1') as fichero_csv:
    lector = csv.reader(fichero_csv)
    next(lector, None)
    importe_total = 0
    for linea in lector:
        importe_str = linea[2]
        importe = float(importe_str)
        print("Importe:", importe)
        importe_total = importe_total + importe
        
    print("Importe Total:", importe_total)
```

Si queremos leer nuestro fichero .csv como un diccionario de datos, lo podemos hacer de la siguiente forma:

```python
with open(path, encoding = 'latin1') as fichero_csv:
    dict_lector = csv.DictReader(fichero_csv)
    asocs = {}
    for linea in dict_lector:
        centro = linea['ASOCIACION']
        subvencion = float(linea['SUBVENCION EN EUROS'])
        if centro in asocs:
            asocs[centro] = asocs[centro] + subvencion
        else:
            asocs[centro] = subvencion
    print(asocs)
```

### Escritura CSV 

```python
path1 = 'C:/Users/USUARIO DELL/Documents/Python Scripts/ANALISIS_DATOS/LECTURA_FICHEROS/subvenciones-educacion2015.csv'
path2 = 'C:/Users/USUARIO DELL/Documents/Python Scripts/ANALISIS_DATOS/LECTURA_FICHEROS/subvenciones-educacion2015_modi.csv'
```

```python
import csv
```

```python
with open (path1, encoding = 'latin1') as archivo_lectura, open(path2, 'w', encoding = 'latin1') as archivo_escritura:
    dict_lector = csv.DictReader(archivo_lectura)
    campos = dict_lector.fieldnames + ['JUSTIFICACION REQUERIDA', 'JUSTIFIACACION RECIBIDA']
    escritor = csv.DictWriter(archivo_escritura, fieldnames = campos)
    escritor.writeheader()
    for linea in dict_lector:
        if float(linea['SUBVENCION EN EUROS']) > 300:
            linea['JUSTIFICACION REQUERIDA'] = 'SI'
        else:
            linea['JUSTIFICACION REQUERIDA'] = 'NO'
        linea['JUSTIFIACACION RECIBIDA'] = 'NO'
        escritor.writerow(linea)
```
