# Request para Extracción de Contenido de Página 12 

```python
# Importamos la biblioteca Requests
import requests
```

```python
# Realizamos conexión con Requests
url = "https://www.pagina12.com.ar"
p12 = requests.get(url)
p12.status_code
```

```python
# Imprimimos el contenido de la página
print(p12.text)
```

```python
# Imprimimos el contenido en caso de no ser un archivo de texto
p12.content
```

```python
# Obtenemos los headers de la página
p12.headers
```

```python
# Obtenemos los headers de la página
p12.request.headers
```

```python
# Obtenemos el método utilizado
p12.request.method
```
